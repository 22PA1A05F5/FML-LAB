# **Activation Function - Sigmoid**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : Sigmoid')
plt.show()
```

# **Activation Function-tanh**

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis("tight")
plt.title("Activation Function: tanh")
plt.show()
```
