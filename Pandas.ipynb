# **SIMPLE CONDITION**
**DATA_FRAME["COL_NAME"]>/</==VALUE**

```python
import pandas as pd
```

```python
dataset=pd.read_excel("/content/drive/MyDrive/Student marks/Student Marks.csv.xlsx")
```

```python
dataset
```

```python
dataset.loc[dataset["Python"]>90] # Python greater than 90 marks will be displayed
```

```python
dataset.loc[dataset["Java"]<80] #less than 80 marks #simple condition
```

```python
dataset.loc[dataset["C Language"]>90 & (dataset["Python"]<85)] #compound condition
```

```python
dataset.loc[dataset["Name of the Student"].str.contains("n")] #wherever there is n,those all names are displayed
```

```python
dataset.loc[dataset["Name of the Student"].str.startswith("h")]
```

```python
dataset
```

```python
dataset.loc[dataset["Name of the Student"].str.startswith("H")]
```

```python
dataset.loc[dataset["Name of the Student"].str.startswith("R")]
```
